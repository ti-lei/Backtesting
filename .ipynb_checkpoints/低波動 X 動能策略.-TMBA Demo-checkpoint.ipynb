# 讀入套件

```python
from preprocess import Data
from backtest import Backtest

import os
from operators import lag
```

# 輸入資料

```python
# 建立Data物件
vol = Data(
    data_name='vol',
    data_frequency='day',
    market_size_threshold=200
)

close_price = Data(
    data_name='收盤價_未還原_已格式化',
    data_frequency='day',
    market_size_threshold=200
)
```

# 定義篩選條件、計算公式

```python
from operators import lag

def calculate_days_250_return(series):
    '''
    計算近250日報酬
    '''
    return (
        series / lag(series, 250) -1
    )

def bigger_than_threshold(series):
    '''
    回傳大於等於0.2者
    '''
    return series >= 0.1
```

# 加入濾網、篩選條件、計算自定義數值分析
範例策略：每月最後一日衡量，波動度最低20%者，且近250日報酬大於0.2。

```python
vol.mutate_data(
    None, 
    rebalance_frequency='month',
    date_order_ascending=False,
    measure_on_nth_date=1
)

close_price.mutate_data(
    calculate_days_250_return,
    rebalance_frequency='month',
    date_order_ascending=False,
    measure_on_nth_date=1
)
```

```python
vol.select(rank='bottom', ratio=0.2)
vol.pipe_to(close_price).add_filter(bigger_than_threshold)
```

# 開始回測

```python
signal_list = [
    close_price 
]

backtesting = Backtest(
    signal_list,
    start_date='2010/1/1',
    end_date='2019/6/30',
    long_or_short='long'
)
backtesting.run()
```
