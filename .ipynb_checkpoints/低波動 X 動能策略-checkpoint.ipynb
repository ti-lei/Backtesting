```python
# 建立Data物件
# vol = Data(
#     data_name='vol',
#     data_frequency='day',
#     market_size_threshold=200
# )

close_price = Data(
    data_name='close_price',
    data_frequency='day',
    market_size_threshold=200
)

close_price_2 = Data(
    data_name='close_price',
    data_frequency='day',
    market_size_threshold=200
)
```

```python
close_price_2.data = close_price_2.data / close_price_2.data.shift(1, axis=1) -1

close_price_2.data = close_price_2.data.rolling(250, axis=1).std()
```

# 定義篩選條件、計算公式

```python
from operators import lag

def calculate_days_60_return(series):
    '''
    計算近250日報酬
    '''
    return (
        series / lag(series, 250) -1
    )

def bigger_than_threshold(series):
    '''
    回傳大於等於0.2者
    '''
    return series > 0.05
```

# 加入濾網、篩選條件、計算自定義數值分析

```python
close_price_2.mutate_data(
    None, 
    rebalance_frequency='month',
    date_order_ascending=False,
    measure_on_nth_date=1
)

close_price.mutate_data(
    calculate_days_60_return,
    rebalance_frequency='month',
    date_order_ascending=False,
    measure_on_nth_date=1
)

# close_price_2.select(rank='bottom', ratio=0.2)
# close_price.add_filter(bigger_than_threshold)

close_price_2.select(rank='bottom', ratio=0.2)
close_price_2.pipe_to(close_price).add_filter(bigger_than_threshold)
```

# 開始回測

```python
signal_list = [
    close_price 
]

backtesting = Backtest(
    signal_list,
    start_date='2010/1/1',
    end_date='2019/8/30',
    long_or_short='long'
)
backtesting.run()
```

# 單期損益累加

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
trade_result = backtesting_result.trade_result

entry = trade_result[trade_result.date_entry_or_exit == 1][
    ['holding_date', 'trading_price', 'code']
]
entry.columns = entry.columns + '_in'
entry.reset_index(drop=True, inplace=True)


exit = trade_result[trade_result.date_entry_or_exit == -1][
     ['holding_date', 'trading_price', 'code']
]
exit.columns = exit.columns +'_out'
exit.reset_index(drop=True, inplace=True)


result = pd.concat([entry, exit], axis=1)
```

```python
result['profit'] = (
    result.trading_price_out - result.trading_price_in - 
    result.trading_price_out*0.004425 -
    result.trading_price_in*0.001425
) * 1000
```
